```python
import whisper
from whisper.utils import get_writer

model = whisper.load_model("turbo")

audio = "/content/drive/MyDrive/karaoke_output/vocals.wav"
result = model.transcribe(audio,language="Japanese", word_timestamps=True)

output_directory = "/content/drive/MyDrive/karaoke_output/"

# Save as an SRT file
srt_writer = get_writer("srt", output_directory)
srt_writer(result, audio)
```

```python
import subprocess
!sudo apt-get update
!sudo apt-get install -y fonts-noto-cjk
```

```python
def lyrics_transpose():
    input_video = "/content/drive/MyDrive/karaoke_output/videoplayback_karaoke.mp4"
    input_lyrics = "/content/drive/MyDrive/karaoke_output/vocals.srt"
    output_video = "/content/drive/MyDrive/karaoke_output/final.mp4"

    cmd = [
        "ffmpeg",
        "-i", input_video,
        "-vf", f"subtitles='{input_lyrics}':force_style='FontName=Noto Sans CJK JP,FontSize=24,PrimaryColour=&HFFFFFF&'",
        "-c:a", "copy",   # keep audio as-is
        output_video
    ]

    result = subprocess.run(cmd, capture_output=True, text=True)

    if result.returncode == 0:
        print("Subtitles successfully added!")
    else:
        print("Error:", result.stderr)

lyrics_transpose()
```
